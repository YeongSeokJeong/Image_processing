```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

## 과제 1

```python
img_file = np.zeros((1280, 1920, 3), dtype = np.int32) # numpy zeros함수를 통해 (1280, 1920, 3) 형태의 ARRAY선언
img_file[1280//7 * 3 : 1280 // 7 * 4, 1920//6 :1920//6 *5, :] = 255 # 먼저 십자가의 가로 줄을 만들기 위한 인덱싱
img_file[1280//7: 1280//7 * 6, 1920//12 * 5:1920//12 * 7, :] =  255 # 십자가의 세로줄을 만들기 위한 인덱싱
plt.imshow(img_file)# 화면 출력
plt.show()
```

```python
a_img = cv2.imread('cat.jpg')
plt.imshow(a_img)
plt.show()
```

```python
print('max :', np.max(a_img), '\tmin :',np.min(a_img))# numpy 함수를 통한 결과 추출
```

```python
max_br, min_br = np.max(a_img), np.min(a_img) # 최대값과 최솟값 저장
v_img = (a_img - min_br)/(max_br -  min_br) * 255 # 화소를 변환하기 위한 연산
v_img = v_img.astype(np.int32) # int 형으로 형을 변환해줌
print('max : ', np.max(v_img), '\tmin :', np.min(v_img))
```

```python
fig = plt.figure(figsize = (10,10)) # figure 선언
ax = fig.add_subplot(1,2,1) # 1행 2열의 첫번째 plot추가
ax = plt.title("Image A") # 타이틀 선언
ax = plt.imshow(a_img) # 이미지 추가

ax = fig.add_subplot(1,2,2)# 1행 2열의 두번째 plot추가
ax = plt.title("Image B") # 타이틀 선언
ax = plt.imshow(v_img, vmin = 0, vmax = 255) # 이미지 추가
```

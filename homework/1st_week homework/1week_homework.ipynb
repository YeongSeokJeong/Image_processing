```python
triangle_size = 5 #    
for i in range(1, triangle_size + 1): # i  [1,2,3,4,5]   
    for _ in range(5 - i): # 5-i    
        print(' ', end = '')
    for _ in range(2 * i - 1): # 2i-1 *  
        print('*', end = '')
    print() #    
```

```python
import sys
print('-----------------------')
print('1.\n2.\n3.\n4.\n5.')
print('-----------------------')
input_num = 0
while True:
    input_num = int(input(' :'))
    if input_num >= 1 and input_num < 5:
        break
    if input_num == 5:# 5    
        sys.exit(0)
while True:
    num = input("0   2 .:").split(',')#  ,  .
    try:#  2     
        num1, num2 = int(num[0]), int(num[1])
    except IndexError: # 
        print("2  ")
        continue
    if num1>0 and num2>0:#     
        break
#    
if input_num == 1:
    print(num1, '+' , num2, '=', num1+num2)
elif input_num == 2:
    print(num1, '-' , num2, '=', num1-num2)
elif input_num == 3:
    print(num1, '*' , num2, '=', num1*num2)
elif input_num == 4:
    print(num1, '/' , num2, '=', num1/num2)
```
